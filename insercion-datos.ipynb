### Inserción de los datos para cuartos

```python
cuartos = pd.read_csv('./datos captura.xlsx - CUARTOS.csv')
```

```python
columnas_cuartos = ['cuarto', 'inicio estadia', 'fin estadia']
```

```python
cuartos.head()
```

```python
df_cuartos = cuartos[['cuarto', 'inicio estadia', 'fin estadia']].rename(columns={'cuarto': 'numero_cuarto', 'inicio estadia': 'inicio_estadia', 'fin estadia': 'fin_estadia'})
```

```python
def convertir_fecha(fecha):
    try:
        return pd.to_datetime(fecha, format='%d/%m/%Y').strftime('%Y-%m-%d')
    except ValueError:
        return pd.to_datetime(fecha, format='%d/%m/%y').strftime('%Y-%m-%d')

# Convertir el formato de las fechas de DD/MM/YY a YYYY-MM-DD utilizando la función personalizada
df_cuartos['inicio_estadia'] = df_cuartos['inicio_estadia'].apply(convertir_fecha)
df_cuartos['fin_estadia'] = df_cuartos['fin_estadia'].apply(convertir_fecha)
```

```python
df_cuartos
```

```python
# Consultar la base de datos para obtener los paciente_id correspondientes a los pacientes en el archivo CSV
pacientes_ids = pd.read_sql("SELECT paciente_id FROM pacientes", motor_db)

# Agregar los paciente_id al DataFrame de cuartos
df_cuartos['paciente_id'] = pacientes_ids['paciente_id']
```

```python
# Insertar los datos en la tabla cuartos
df_cuartos.to_sql(name='cuartos', con=motor_db, if_exists='append', index=False)
```

### Insertar cuentas pacientes detalles

```python
print(df_csv['cuenta1'].head())
print(df_csv['cuenta2'].head())
```

```python


# Eliminar todos los caracteres que no sean dígitos numéricos de las columnas 'cuenta1' y 'cuenta2'
df_csv['cuenta1'] = df_csv['cuenta1'].str.replace(r'\D', '', regex=True)
df_csv['cuenta2'] = df_csv['cuenta2'].str.replace(r'\D', '', regex=True)

# Convertir las columnas a double
df_csv['cuenta1'] = df_csv['cuenta1'].astype(float)
df_csv['cuenta2'] = df_csv['cuenta2'].astype(float)

# Eliminar los dos últimos ceros de las columnas
df_csv['cuenta1'] = df_csv['cuenta1'] / 100
df_csv['cuenta2'] = df_csv['cuenta2'] / 100
```

```python
df_csv['cuenta1']
```

```python
# Sumar las columnas 'cuenta1' y 'cuenta2' y colocar el resultado en la columna 'costo_total'
df_csv['costo_total'] = df_csv['cuenta1'] + df_csv['cuenta2']

# Sumar las columnas 'cantidad1' y 'cantidad2' y colocar el resultado en la columna 'cantidad'
df_csv['cantidad'] = df_csv['cantidad1'] + df_csv['cantidad2']
```

```python
df_csv.head()
```

```python

# Realizar la consulta SQL para obtener los cuentas_pacientes_id de la tabla cuentas_pacientes
consulta_cuentas_pacientes_id = "SELECT cuentas_pacientes_id FROM cuentas_pacientes"
df_cuentas_pacientes_id = pd.read_sql(consulta_cuentas_pacientes_id, motor_db)

# Agregar los cuentas_pacientes_id al DataFrame actual
df_csv['cuentas_pacientes_id'] = df_cuentas_pacientes_id
```

```python
# Crear un nuevo DataFrame con las columnas especificadas
df_nuevo = df_csv[['cuentas_pacientes_id', 'Descripcion1', 'cantidad', 'costo_total']].copy()

# Renombrar las columnas
df_nuevo.rename(columns={'Descripcion1': 'detalle'}, inplace=True)
```

```python
df_nuevo.head()
```

```python
df_nuevo.to_sql('cuentas_pacientes_detalles', con=motor_db, if_exists='append', index=False)
```
